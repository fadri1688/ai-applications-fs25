# Feature Engineering

This notebook create new features and mesures the the perfomance change in the model. The goal is to create features to increase the model performance.

```python
# Libraries
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

### Import the data

```python
# Read the data to a pandas data frame
df = pd.read_csv('original_apartment_data_analytics_hs24_with_lat_lon.csv', sep=',', encoding='utf-8')
# Get number of rows and columns
df.shape
```

```python
df.columns
```

## Model Performance

```python
def model_performance(features, df, random_forest_model = RandomForestRegressor(random_state=42)):
    df = df.sample(frac=1, random_state=42)
    X, y = df[features], df['price']
    scores = cross_val_score(random_forest_model, X, y, scoring="neg_root_mean_squared_error", cv=5)
    print('CV results RMSE:', np.round(scores))
    print('Mean RMSE:', np.mean(np.round(scores, 0)))
```

```python
features = ['rooms', 'area', 'pop', 'pop_dens', 'frg_pct', 'emp', 'tax_income']
model_performance(features, df)
```

## Data Cleaning
- Remove apartments with empty columns
- Remove apartments which price exceeds 6000.-
- Remove apartments which price is lower than 1000.-

```python
print('Total apartment before data cleaning:', len(df))

# Remove missing values
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

# Remove some 'extreme' values
df = df.loc[(df['price'] >= 750) & 
            (df['price'] <= 8000)]

print('Total apartment after data cleaning:', len(df))
```

### Luxurious One Hot Encoding
One-Hot Encoding is a technique used in machine learning to convert categorical variables into a binary (0/1) format. Each unique category is represented as a separate column, and a 1 is assigned if that category is present, otherwise, it's 0.

Example:

| description_raw                                         | (ATTIKA) | (EXKLUSIV) | (LOFT) | (LUXURIÖS) | (POOL) | (SEESICHT) | (NONE) |
|---------------------------------------------------------|----------|------------|--------|------------|--------|------------|--------|
| Dieses Apartment ist sehr LUXURIÖS mit POOL.           | 0        | 0          | 0      | 1          | 1      | 0          | 0      |
| Ein tolles LOFT mit SEESICHT und EXKLUSIVEM Design.    | 0        | 1          | 1      | 0          | 0      | 1          | 0      |
| Normale Wohnung ohne Extras.                           | 0        | 0          | 0      | 0          | 0      | 0          | 1      |
| Schöne Aussicht, aber keine besonderen Merkmale.       | 0        | 0          | 0      | 0          | 0      | 0          | 1      |

### Zusatzt Feature

```python
import numpy as np
from math import radians, cos, sin, asin, sqrt

# Funktion zur Berechnung der Entfernung zwischen zwei Koordinatenpunkten
def haversine(lon1, lat1, lon2, lat2):
    """
    Berechnet die Entfernung in Kilometern zwischen zwei Punkten 
    auf der Erde (anhand von Längen-/Breitengraden)
    """
    # Umwandlung in Radians
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    
    # Haversine-Formel
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    km = 6371 * c  # Erdradius in Kilometern
    return km

# Liste der wichtigsten S-Bahn und Verkehrsknotenpunkte in Zürich
transport_hubs = [
    ["Zürich HB", 47.3778, 8.5403],
    ["Zürich Stadelhofen", 47.3665, 8.5486],
    ["Zürich Oerlikon", 47.4111, 8.5438],
    ["Zürich Altstetten", 47.3911, 8.4889],
    ["Zürich Wiedikon", 47.3708, 8.5231],
    ["Zürich Hardbrücke", 47.3861, 8.5186],
    ["Zürich Enge", 47.3642, 8.5306],
    ["Zürich Wollishofen", 47.3447, 8.5327],
    ["Zürich Tiefenbrunnen", 47.3519, 8.5681],
    ["Zürich Seebach", 47.4242, 8.5475],
    ["Winterthur", 47.4996, 8.7241],
    ["Winterthur Grüze", 47.4994, 8.7494],
    ["Winterthur Töss", 47.4878, 8.6987],
    ["Winterthur Seen", 47.4793, 8.7674],
    ["Uster", 47.3474, 8.7201],
    ["Wetzikon", 47.3281, 8.7974],
    ["Pfäffikon ZH", 47.3662, 8.7827],
    ["Hinwil", 47.2976, 8.8432],
    ["Dübendorf", 47.3973, 8.6186],
    ["Wallisellen", 47.4094, 8.5915],
    ["Dietlikon", 47.4154, 8.6228],
    ["Opfikon", 47.4247, 8.5721],
    ["Kloten", 47.4515, 8.5849],
    ["Flughafen Zürich", 47.4506, 8.5615],
    ["Bülach", 47.5191, 8.5373],
    ["Embrach-Rorbas", 47.5005, 8.5934],
    ["Dietikon", 47.4051, 8.4012],
    ["Schlieren", 47.3961, 8.4477],
    ["Regensdorf-Watt", 47.4387, 8.4698],
    ["Meilen", 47.2704, 8.6452],
    ["Horgen", 47.2552, 8.6006],
    ["Thalwil", 47.2916, 8.5641]
]

# Funktion zur Berechnung der Entfernung zur nächsten Station
def distance_to_nearest_hub(lat, lon, hubs):
    """
    Berechnet die Entfernung zum nächsten Verkehrsknotenpunkt
    """
    distances = []
    for hub in hubs:
        hub_name, hub_lat, hub_lon = hub
        distance = haversine(lon, lat, hub_lon, hub_lat)
        distances.append(distance)
    
    return min(distances)  # Kürzeste Entfernung zurückgeben

# Anwendung auf den Datensatz
df['distance_to_transport'] = df.apply(
    lambda row: distance_to_nearest_hub(row['lat'], row['lon'], transport_hubs), 
    axis=1
)

# Optional: Den Namen des nächsten Knotenpunkts speichern
def nearest_hub_name(lat, lon, hubs):
    """
    Gibt den Namen des nächsten Verkehrsknotenpunkts zurück
    """
    distances = []
    for hub in hubs:
        hub_name, hub_lat, hub_lon = hub
        distance = haversine(lon, lat, hub_lon, hub_lat)
        distances.append((distance, hub_name))
    
    return min(distances, key=lambda x: x[0])[1]  # Name des nächsten Knotenpunkts

df['nearest_transport_hub'] = df.apply(
    lambda row: nearest_hub_name(row['lat'], row['lon'], transport_hubs), 
    axis=1
)

# Zusammenfassung der neuen Features zeigen
print(df[['distance_to_transport', 'nearest_transport_hub']].describe())

# Korrelation mit dem Preis überprüfen
print(f"Korrelation zwischen Entfernung zum Verkehrsknotenpunkt und Preis: {df['distance_to_transport'].corr(df['price'])}")

# Das neue Feature zum Modell hinzufügen und Performance testen
features_with_transport = features + ['distance_to_transport']
model_performance(features_with_transport, df)

print("-------------------------------------------------")
#Kategorien erstellen
# Create distance category feature
def categorize_distance(distance):
    if distance < 1:
        return "Walking distance"
    elif distance < 5:
        return "Short commute"
    elif distance < 15:
        return "Medium commute"
    else:
        return "Long commute"

# Apply the function to create a new categorical column
df['transport_distance_category'] = df['distance_to_transport'].apply(categorize_distance)

# Check the distribution of categories
print(df['transport_distance_category'].value_counts())

# Map categories to numbers (preserving the order of distances)
distance_categories = df['transport_distance_category'].values.reshape(-1, 1)
ordinal_encoder = OrdinalEncoder(categories=[['Walking distance', 'Short commute', 'Medium commute', 'Long commute']])
df['transport_distance_encoded'] = ordinal_encoder.fit_transform(distance_categories)

# Preview the result
print(df[['distance_to_transport', 'transport_distance_category', 'transport_distance_encoded']].head(10))

# Test the model performance with both continuous and categorical distance features
features_with_transport_cat = features + ['distance_to_transport', 'transport_distance_encoded']
model_performance(features_with_transport_cat, df)

# Alternatively, you can also create binary dummy variables (one-hot encoding)
# This might capture non-linear relationships better
transport_dummies = pd.get_dummies(df['transport_distance_category'], prefix='transport')
df = pd.concat([df, transport_dummies], axis=1)

# Test with one-hot encoded features
features_with_transport_one_hot = features + ['distance_to_transport'] + transport_dummies.columns.tolist()
model_performance(features_with_transport_one_hot, df)
```
